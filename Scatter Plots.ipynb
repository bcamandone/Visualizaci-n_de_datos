## Scatter Plots

```python
#Importamos Librer√≠as
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

#Dataset
penguins = sns.load_dataset('penguins')
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
penguins.head()
```

```python
# Simple Scatterplot
sns.scatterplot(data=tips, x='total_bill', y='tip', s=150, alpha=0.7)
plt.title('Simple Scatterplot')
plt.show();
```

```python
# Scatterplot with Hue
sns.scatterplot(data=penguins, x='bill_depth_mm', y='flipper_length_mm', s=150, alpha=0.7,
               hue='species', palette='viridis')
plt.title('Scatterplot with Hue')
plt.show()  
```

```python
# Scatterplot with Hue & Size
sns.scatterplot(data=tips, x='total_bill', y='tip', sizes=(40, 400), alpha=.5,
              palette='viridis', hue='time', size='size')
plt.title('Scatterplot with Hue & Size')
plt.show()
```

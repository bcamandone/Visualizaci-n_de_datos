## Histogramas

```python
#Importamos Librer√≠as
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

#Dataset
titanic = sns.load_dataset('titanic')
penguins = sns.load_dataset('penguins')
```

```python
titanic.head()
```

```python
penguins.head()
```

```python
# Histograma 
sns.histplot(data=titanic, x='age', kde=True)
plt.title('Histograma')
plt.show();
```

```python
# Histograma con multiples variables
sns.histplot(data= penguins[['bill_length_mm','bill_depth_mm']])
plt.title('Histograma multiples variables')
plt.show();
```

```python
#Stacked Histogram
sns.histplot(penguins, x='flipper_length_mm', hue='species', multiple='stack',
             linewidth=0.5)
plt.title('Histograma apilado')
plt.show()
```
